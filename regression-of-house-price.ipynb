## **Function: grab_col_names

```python
def grab_col_names(dataframe, cat_th=20, car_th=25):
    """

    Veri setindeki kategorik, numerik ve kategorik fakat kardinal değişkenlerin isimlerini verir.
    Not: Kategorik değişkenlerin içerisine numerik görünümlü kategorik değişkenler de dahildir.

    Parameters
    ------
        dataframe: dataframe
                Değişken isimleri alınmak istenilen dataframe
        cat_th: int, optional
                numerik fakat kategorik olan değişkenler için sınıf eşik değeri
        car_th: int, optinal
                kategorik fakat kardinal değişkenler için sınıf eşik değeri

    Returns
    ------
        cat_cols: list
                Kategorik değişken listesi
        num_cols: list
                Numerik değişken listesi
        num_but_car: list
                 Numerik görünümlü kardinal değişken listesi
        cat_but_car: list
                Kategorik görünümlü kardinal değişken listesi

    Examples
    ------
        import seaborn as sns
        df = sns.load_dataset("iris")
        print(grab_col_names(df))


    Notes
    ------
        cat_cols + num_cols + cat_but_car = toplam değişken sayısı
        num_but_cat cat_cols'un içerisinde.
        Return olan 3 liste toplamı toplam değişken sayısına eşittir: cat_cols + num_cols + cat_but_car = değişken sayısı

    """

    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]
          
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}\n")
    print(f'cat_cols: {len(cat_cols)}; {cat_cols}\n')
    print(f'num_cols: {len(num_cols)}; {num_cols}\n')
    print(f'cat_but_car: {len(cat_but_car)}; {cat_but_car}\n')
    print(f'num_but_cat: {len(num_but_cat)}; {num_but_cat}\n')

    return cat_cols, num_cols, cat_but_car, num_but_cat
```

## 2.1 MISSING VALUES

> *missing_values_table* funcsion is for finding the missing value in any variable. In this sample, any missing value not means missing. it means the house has not the variable. For example, if the variable *"GarageType"* has NaN value, it means the hause has not a garage. For this sample, suitable missing values are replaced with *No* (No_Garage etc.)

## **Function: missing_values_table

```python
def missing_values_table(dataframe, na_name=False, plot=False):
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]

    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False)
    ratio = (dataframe[na_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False)
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio'])

    print(missing_df, end="\n")

    if plot and not missing_df.empty:
        plt.figure(figsize=(12, 6))
        ax = missing_df['n_miss'].plot(kind='bar', color='skyblue', alpha=0.8, edgecolor='black')

        # Oranları çubukların üstüne ekleme
        for idx, value in enumerate(missing_df['n_miss']):
            ratio_value = missing_df['ratio'].iloc[idx]
            ax.text(idx, value + 1, f'{ratio_value:.1f}%', ha='center', va='bottom', fontsize=8, color='black')

        # Başlık ve eksen bilgileri
        plt.title('Missing Values by Variables', fontsize=14)
        plt.ylabel('Number of Missing Values', fontsize=12)
        plt.xlabel('Variables', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    if na_name:
        return na_columns
```

```python
def missing_vs_target(dataframe, target, na_columns, plot=False):
    temp_df = dataframe.copy()

    for col in na_columns:
        temp_df[col + '_NA_FLAG'] = np.where(temp_df[col].isnull(), 1, 0)

    na_flags = temp_df.loc[:, temp_df.columns.str.contains("_NA_")].columns

    for col in na_flags:
      
        
        group_data = temp_df.groupby(col)[target].agg(["mean", "count"])
        group_data["ratio"] = group_data["count"] / group_data["count"].sum()
        print(pd.DataFrame({"TARGET_MEAN": temp_df.groupby(col)[target].mean(),
                            "Count": temp_df.groupby(col)[target].count(),
                            "Ratio": group_data["ratio"]}), end="\n\n\n")
```

## 2.1 FILLING NAN VALUES

```python
# FILL NAN VALUES WITH NO, IF HOUSE HAS NOT A FEATURE

has_words = ['GARAGE', "POOL", "FENCE", "FIREPLACE", "MISC", "ALLEY", "LOTFRONTAGE"]
hascol = [col for col in df.columns if any(word in col.upper() for word in has_words)]

df[hascol] = df[hascol].apply(
    lambda x: x.fillna('No' if x.dtype == 'object' else 0))

# FILL NAN VALUES WITH MODE, IF VARIBALE IS OBJECT TYPE
nan_cols_fill_mode = [col for col in df.columns if df[col].dtype == 'object' and df[col].isna().any()]
nan_cols_fill_mode = [col for col in nan_cols_fill_mode if col not in hascol]

df[nan_cols_fill_mode] = df[nan_cols_fill_mode].fillna(df[nan_cols_fill_mode].mode().iloc[0])

# FILL NAN VALUES WITH MEDIAN, IF VARIBALE IS NOT OBJECT TYPE

nan_cols_fill_median = [col for col in df.columns if df[col].dtype != 'object' and df[col].isna().any()]
nan_cols_fill_median = [col for col in nan_cols_fill_median if col not in ["SALEPRICE"]]

df[nan_cols_fill_median] = df[nan_cols_fill_median].fillna(df[nan_cols_fill_median].median())
```

```python
# Creating New Features

df["NEW_BUILT_AGE"] = abs(df["YRSOLD"] - df["YEARBUILT"])
df["NEW_T_PORCH_AREA"] = df["OPENPORCHSF"] + df["3SSNPORCH"] + df["ENCLOSEDPORCH"] + df["SCREENPORCH"]
df["NEW_T_BUILT_AREA"] = df["TOTALBSMTSF"] + df["1STFLRSF"] + df["2NDFLRSF"] + df["GARAGEAREA"] + df["POOLAREA"] + df["NEW_T_PORCH_AREA"] + df["WOODDECKSF"]
df["NEW_T_AREA_R"] = df["TOTALBSMTSF"] + df["1STFLRSF"] + df["2NDFLRSF"] + (df["GARAGEAREA"] / 3) + (df["POOLAREA"] / 5) + (df["NEW_T_PORCH_AREA"] / 5) + (df["WOODDECKSF"] / 3)
df["NEW_TOTAL_UFNS_AREA"]=df[[col for col in df.columns if 'UNF' in col]].sum(axis=1)
df["NEW_UFNS_AREA_FACTOR"]=df["NEW_TOTAL_UFNS_AREA"]/(df["NEW_T_BUILT_AREA"]+1)
df["NEW_BUILT_LOT_RATIO"]=df["NEW_T_BUILT_AREA"]/df["LOTAREA"]
df["NEW_AGE_FACTOR"] = 1 / (df["NEW_BUILT_AGE"] + 1)
df["NEW_LOT_AREA_FACTOR"] = df["LOTFRONTAGE"]/df["LOTAREA"]

df["NEW_LOT_SHAPE_FACTOR"] = [1 if lots in ["Rare"]
                            else 2 if lots in ["IR1"]
                            else 3 for lots in df["LOTSHAPE"]]

df["NEW_LOT_CONF_FACTOR"] = [1 if lotc in ["Rare"]
                            else 2 if lotc in ["CulDSac"]
                            else 3 if lotc in ['FR2']
                            else 4 if lotc in ['Inside']
                            else 5 for lotc in df["LOTCONFIG"]]

df["NEW_BLD_TYPE_FACTOR"] = [1 if buty in ["1Fam"]
                            else 2 if buty in ["2fmCon"]
                            else 3 if buty in ['TwnhsE']
                            else 4 if buty in ['Twnhs']
                            else 5 for buty in df["BLDGTYPE"]]

df["NEW_HOU_STL_FACTOR"] = [1 if stl in ["1Story"]
                        else 2 if stl in ['1.5Fin', '1.5Unf']
                        else 3 if stl in ['SFoyer', 'SLvl']
                        else 4 if stl in ["2Story"]
                        else 5 for stl in df["HOUSESTYLE"]]

df["NEW_TYPE_STYLE_FACTOR'"] = df['NEW_BLD_TYPE_FACTOR'] * df['NEW_HOU_STL_FACTOR']

df["NEW_CLASS_FACTOR"] = [1 if clas in [20, 30, 40]
                        else 2 if clas in [45, 50, 60, 70, 160, 190]
                        else 3 if clas in [120, 150]
                        else 4 if clas in [160, 180, 190]
                        else 5 for clas in df["MSSUBCLASS"]]

df["NEW_ZONE_FACTOR"] = [1 if zone in ["C (all)"]
                        else 2 if zone in ["A", "I"]
                        else 3 if zone in ["RL", "RM"]
                        else 4 if zone in ["RH", "RP"]
                        else 5 for zone in df["MSZONING"]]

df["NEW_CLASS_ZONE_FACTOR'"] = df['NEW_CLASS_FACTOR'] * df['NEW_ZONE_FACTOR']
df["NEW_LIV_AREA_FACTOR"] = df["GRLIVAREA"] / df["NEW_T_BUILT_AREA"]
df["NEW_QUALITY_FACTOR"] = (df["OVERALLQUAL"] + df["OVERALLCOND"]) / 2
df["NEW_WET_AREA_FACTOR"] = df["FULLBATH"] + df["HALFBATH"] + df["KITCHENABVGR"]
df['NEW_LOT_FACTOR'] = df['LOTFRONTAGE'] / df['LOTAREA']
```

## 2.3 OUTLIER PROCESES

## **Function: outlier_thresholds

```python
def outlier_thresholds(dataframe, col_name, q1=0.05, q3=0.95):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return round(low_limit, 4), round(up_limit,4)
```

```python
for col in num_cols:
    print(col, outlier_thresholds(df, col, q1=0.05, q3=0.95))
```

## **Function: grab_outliers

```python
def grab_outliers(dataframe, col_name, index=False):
    low, up = outlier_thresholds(dataframe, col_name)

    if dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].shape[0] > 10:
        print(dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].head())
    else:
        print(dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))])

    if index:
        outlier_index = dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].index
        return outlier_index
```

```python
for col in num_cols:
    grab_outliers(df, col)
```

## **Function: check_outlier

```python
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
```

```python
for col in num_cols:
    print(col, check_outlier(df, col))
```

## **Function: replace_with_thresholds

```python
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

## 2.2.1 Rare Encoding

```python
def rare_analyser(dataframe, target, cat_cols):
    for col in cat_cols:
        print(col, ":", len(dataframe[col].value_counts()))
        print(pd.DataFrame({"COUNT": dataframe[col].value_counts(),
                            "RATIO": dataframe[col].value_counts() / len(dataframe),
                            "TARGET_MEAN": dataframe.groupby(col)[target].mean()}), end="\n\n\n")
```

```python
def rare_encoder(dataframe: object, rare_perc: object) -> object:
    temp_df = dataframe.copy()

    rare_columns = [col for col in temp_df.columns if temp_df[col].dtypes == 'O'
                    and (temp_df[col].value_counts() / len(temp_df) < rare_perc).any(axis=None)]

    for var in rare_columns:
        tmp = temp_df[var].value_counts() / len(temp_df)
        rare_labels = tmp[tmp < rare_perc].index
        temp_df[var] = np.where(temp_df[var].isin(rare_labels), 'RARE', temp_df[var])

    return temp_df
```

```python
def evaluate_regression_models(X_train, X_test, y_train, y_test, X, y):
    # Define the regression models
    regression_models = {
        "LinRegn_Model": LinearRegression(),
        "CatBoost_Model": CatBoostRegressor(verbose=0, random_state=24),
        "GBM_Model": GradientBoostingRegressor(random_state=24, verbose=0),
        "XGBM_Model": XGBRegressor(random_state=24, verbosity=0),
        "LGBM_Model": LGBMRegressor(random_state=24, verbose=0),
        "KNN_Model": KNeighborsRegressor(),
        "CART_Model": DecisionTreeRegressor(random_state=24),
        "RF_Model": RandomForestRegressor(random_state=24, verbose=0)
    }
    
    # Store results and models
    regression_results = []
    trained_models = {}
    
    for name, model in regression_models.items():
        # Train the model
        model.fit(X_train, y_train)
        trained_models[name] = model  # Save the trained model
        
        # Predict on the test set
        y_pred = model.predict(X_test)
        
        # Compute metrics
        MAE = mean_absolute_error(y_test, y_pred)
        MSE = mean_squared_error(y_test, y_pred)
        RMSE = np.sqrt(MSE)
        r2 = r2_score(y_test, y_pred)
        cv_score = np.mean(cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error'))
        
        # Save the metrics
        regression_results.append({
            "Model": name,
            "MAE": MAE,
            "MSE": MSE,
            "RMSE": RMSE,
            "R_SQRT": r2,
            "CV Score (MAE)": -cv_score  # Convert negative CV score to positive
        })
        
     
        
        # Print results
        print(f"\nevaluate_regression_models Result\nRegression Models Comparing Table:\n{results_df}")
    
        # Plot 1: Actual vs Predicted
        plt.figure(figsize=(12, 6))
        plt.scatter(y_test, y_pred, color="blue", alpha=0.6, label="Predicted")
        plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label="Ideal Fit")
        plt.xlabel("Actual Values")
        plt.ylabel("Predicted Values")
        plt.title(f"Actual vs Predicted - {name}")
        plt.legend()
        plt.grid()
        plt.show()
            
        # Plot 2: Residuals Plot
        residuals = y_test - y_pred
        plt.figure(figsize=(12, 6))
        plt.scatter(y_pred, residuals, color="purple", alpha=0.6)
        plt.axhline(y=0, color="red", linestyle="--")
        plt.xlabel("Predicted Values")
        plt.ylabel("Residuals")
        plt.title(f"Residuals Plot - {name}")
        plt.grid()
        plt.show()

  
    # Return the results DataFrame and trained models
    return results_df, trained_models
```

```python
# Evaluate models and find the best one
results_df, trained_models = evaluate_regression_models(X_train, X_test, y_train, y_test, X, y)
```

```python
def optimize_best_model(results_df, trained_models, X_train, y_train, X_test, y_test):
    # Find the model with the highest R_SQRT
    best_model_name = results_df.loc[results_df['RMSE'].idxmin(), 'Model']
    print(f"\n\nBest model based on RMSE: {best_model_name}")
    
    # Retrieve the corresponding trained model
    best_model = trained_models[best_model_name]
    
    # Define hyperparameter grid based on the model type
    param_grid = {}
    if best_model_name == "CatBoost_Model":
        param_grid = {
            "depth": [4, 6, 8, 10],
            "learning_rate": [0.01, 0.05, 0.1],
            "iterations": [100, 500, 1000]
        }
    elif best_model_name == "GBM_Model":
        param_grid = {
            "n_estimators": [100, 200, 300],
            "max_depth": [3, 4, 5],
            "learning_rate": [0.01, 0.05, 0.1],
            "subsample": [0.8, 1.0]
        }
    elif best_model_name == "XGBM_Model":
        param_grid = {
            "n_estimators": [100, 200, 300],
            "max_depth": [3, 4, 5],
            "learning_rate": [0.01, 0.05, 0.1],
            "colsample_bytree": [0.7, 0.8, 1.0]
        }
    elif best_model_name == "LGBM_Model":
        param_grid = {
            "n_estimators": [100, 200, 300],
            "num_leaves": [20, 31, 40],
            "learning_rate": [0.01, 0.05, 0.1]
        }
    elif best_model_name == "RF_Model":
        param_grid = {
            "n_estimators": [100, 200, 300],
            "max_depth": [10, 20, None],
            "min_samples_split": [2, 5, 10]
        }
    elif best_model_name == "CART_Model":
        param_grid = {
            "max_depth": [3, 5, 10, None],
            "min_samples_split": [2, 5, 10],
            "min_samples_leaf": [1, 2, 4]
        }
    elif best_model_name == "KNN_Model":
        param_grid = {
            "n_neighbors": [3, 5, 7, 9],
            "weights": ["uniform", "distance"],
            "metric": ["euclidean", "manhattan"]
        }
    
    # Perform hyperparameter optimization using RandomizedSearchCV
    hpo_grid = RandomizedSearchCV(
        estimator=best_model,
        param_distributions=param_grid,
        scoring="r2",
        n_iter=20,
        cv=5,
        verbose=1,
        random_state=24,
        n_jobs=-1
    )
    
    # Fit the optimizer on training data
    hpo_grid.fit(X_train, y_train)
    
    # Retrieve the best parameters and model
    best_params = hpo_grid.best_params_
    final_best_model = hpo_grid.best_estimator_
    
    # Make predictions using the optimized model
    final_pred = final_best_model.predict(X_test)
    
    # Evaluate the optimized model on the test set
    MAE = mean_absolute_error(y_test, final_pred)
    MSE = mean_squared_error(y_test, final_pred)
    RMSE = np.sqrt(MSE)
    r2 = r2_score(y_test, final_pred)

     # Plotting Actual vs Predicted values
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, final_pred, color='blue', alpha=0.6, label='Predictions')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Ideal Fit')
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"Actual vs Predicted Values ({best_model_name})")
    plt.legend()
    plt.grid()
    plt.show()
    
    # Residuals Plot
    residuals = y_test - final_pred
    plt.figure(figsize=(10, 6))
    plt.scatter(final_pred, residuals, color='purple', alpha=0.6)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.xlabel("Predicted Values")
    plt.ylabel("Residuals")
    plt.title(f"Best Model Residuals Plot ({best_model_name})")
    plt.grid()
    plt.show()
    
    print("\n\nOptimized Model Performance on Test Data:\n")
    print(f"MAE: {MAE:.4f}")
    print(f"MSE: {MSE:.4f}")
    print(f"RMSE: {RMSE:.4f}")
    print(f"R²: {r2:.4f}")
    
    print(f"\nBest parameters for {best_model_name}:\n {best_params}")
    
    return final_best_model, {"MAE": MAE, "MSE": MSE, "RMSE": RMSE, "R2": r2}
```

```python
# Final Prediction and Submission
def generate_submission(final_model, train_data, feature_columns, target_column="SALEPRICE", id_column="ID", log_transformed=False):
    """
    Generate submission DataFrame from the final model.
    
    Args:
        final_model: Trained and optimized regression model.
        train_data: DataFrame containing features and IDs for prediction.
        feature_columns: List of columns used as features for prediction.
        target_column: Name of the target column for submission.
        id_column: Name of the ID column for submission.
        log_transformed: Boolean indicating if the target variable was log-transformed.
    
    Returns:
        submission_df: DataFrame ready for submission.
    """
    # Predict using the final model
    predictions = final_model.predict(train_data[feature_columns])
    
    # Reverse log transformation if required
    if log_transformed:
        predictions = np.exp(predictions)
    
    # Create submission DataFrame
    submission_df = pd.DataFrame({id_column: train_data[id_column], target_column: predictions})
    submission_df[id_column] = submission_df[id_column].astype(int)
    submission_df.reset_index(drop=True, inplace=True)
    
    return submission_df
```
